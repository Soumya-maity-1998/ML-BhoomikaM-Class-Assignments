```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
''' Gradient Descent Method '''

t = np.array([0, 1, 2, 3, 4])
lnn = np.array([6.9, 6.4, 5.8, 5.3, 4.9])

N = len(t) 

m = 0.0; c = 0.0    # Initial guess

eta = 0.1   # Learning Rate 
epochs = 500

cost_history = []

for i in range(epochs):
    hypothesis = m*t + c  
    cost_function = (0.5/N)*np.sum((hypothesis - lnn)**2)
    cost_history.append(cost_function) 

    dm = (1/N)*np.sum((hypothesis - lnn)*t) 
    dc = (1/N)*np.sum(hypothesis - lnn)  

    m = m - eta*dm 
    c = c - eta*dc 


print(m, c)
```

```python
''' Using libraries '''
import numpy as np
from sklearn.linear_model import SGDRegressor

t = np.array([0, 1, 2, 3, 4]).reshape(-1, 1)
lnn = np.array([6.9, 6.4, 5.8, 5.3, 4.9]) 

model = SGDRegressor(loss = 'squared_error', max_iter = 100, learning_rate = 'constant', eta0 = 0.1, random_state = 42)

model.fit(t, lnn) 
c1 = model.intercept_[0]
m1 = model.coef_[0]

print(m1, c1)
```

```python
x = np.linspace(0, 5, 100)

plt.plot(t, lnn, '*', label = 'Experimental Data')
plt.plot(x, m1*x + c, label = 'Fit by SDGRegressor')
plt.plot(x, m*x + c, '-.', label = 'Fit by script')
plt.legend()
plt.grid()
plt.xlabel('Time--->')
plt.ylabel('ln N --->')
plt.title('Decay Constant Calculation')
plt.show()
```
